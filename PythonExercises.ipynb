Exercise 1 : Calculate Simple Interest

```python
# principal = amount of money deposited or borrowed
# rate = rate of interest per year as a percentage
# time = duration in years
# interest = simple interest 
def calculate_interest(principal, rate, time):
    interest = (principal * rate * time) / 100
    return interest
```

```python
# interest = calculate_interest(1000, 5, 2)
# interest = calculate_interest(1500, 4.3, 3)
interest = calculate_interest(500, 10, 0)
print(f"The interest is {interest}")
```

Exercise 2 : Calculate Hypotense

```python
# base = base of the triangle
# height = height of the triangle
# hypotenuse = hypotenuse calculated using the pythagorean theorem
def calculate_hypotenuse(base, height):
    hypotenuse = (base ** 2 + height ** 2) ** 0.5
    return hypotenuse
    
```

```python
# hypotenuse = calculate_hypotenuse(3, 4)
# hypotenuse = calculate_hypotenuse(5, 12)
hypotenuse = calculate_hypotenuse(0, 0)
print(f"Hypotenuse is {hypotenuse}")
```

Exercise 3 : Find distance in 2D

```python
def find_distance_2d(x1, y1, x2, y2):
    distance = ((y2-y1) ** 2 + (x2-x1) ** 2 ) ** 0.5
    return distance
```

```python
distance = find_distance_2d(0, 0, 3, 4)     # Output: 5.0
# distance = find_distance_2d(1, 2, 4, 6)     # Output: 5.0
# distance = find_distance_2d(1, 2, 4, 6)     # Output: 5.0
print(f" Distance in 2d is {distance}")
```
